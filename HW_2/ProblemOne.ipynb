**Imports And Style Configurations**

```
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

plt.style.use('seaborn-white')
```

We define a point p = p1 = p2 = p3 = (px, py), such that when inserted into the functions the following is true:
* It is the minimum of f1
* It is the maximum of f2
* It is neither a minimum or maximum of f3, but the gradient is 0 at p

```
px = 0
py = 0

def f1(x,y):
  return x**2 + y**2

def f2(x,y):
  return -(x**2) - y**2

def f3(x,y):
  return x**3 + y**3
```

<h1>3D Functions</h1>

We define a function that takes three parameters: a function 'f', px, and py;

such that a 3D plot is graphed from these. The graphical settings are specifically made to accomodate f1, f2, and f3.

```
def plot_3d_function (f, px, py):
  # Sets dimensions for plot
  x = np.linspace(-3, 3, 30)
  y = np.linspace(-3, 3, 30)
  X, Y = np.meshgrid(x, y)
  Z = f(X, Y)

  # Configure plot
  fig = plt.figure()
  ax = plt.axes(projection='3d')

  # Plots the surface
  ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
  # Plots the point
  pz = f(px, py)
  ax.plot3D([px], [py], [pz], 'ro', label='Point ('+str(px)+','+str(py)+')')
  ax.legend(loc='upper left', frameon=True, framealpha=1)

  # Sets axis labels and title
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')
  ax.set_title(f.__name__)

  # Sets viewing angle
  ax.view_init(50,-90)
  return
```

Now we plot those functions and get the following:

```
plot_3d_function(f1, px, py)
plot_3d_function(f2, px, py)
plot_3d_function(f3, px, py)
```

<h1>Contour Plots</h1>

Now we can represent those same functions as contour plots. The function takes in the same parameters as the 3D function and is also suited specifically for f1, f2, and f3.

```
def plot_contour_function (f, px, py):
  # Configures the axis
  x = np.linspace(-3, 3, 50)
  y = np.linspace(-3, 3, 50)
  X, Y = np.meshgrid(x, y)
  Z = f(X, Y)

  # COnfigures the plot
  fig = plt.figure()
  ax = plt.axes()

  # Plots the contour
  ax.contour(X, Y, Z, 20, cmap='RdGy')

  # Plots the point (px, py) and configures its legend
  ax.plot([px], [py], 'ro', label='Point ('+str(px)+','+str(py)+')')
  ax.legend(loc='upper left', frameon=True, framealpha=0.8)

  # Sets axis labels and title
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title(f.__name__)
  return
```

Plotting f1, f2, and f3 gives the following results:

```
plot_contour_function(f1, px, py)
plot_contour_function(f2, px, py)
plot_contour_function(f3, px, py)
```

<h1>Density Plots</h1>

Finally, we can express these in terms of density plots, which will be similar to the contour plot function, but gives more detail as to the exact nature of the function itself.

```
def plot_density_function (f, px, py):
  # Configures the axis
  x = np.linspace(-3, 3, 20)
  y = np.linspace(-3, 3, 20)
  X, Y = np.meshgrid(x, y)
  Z = f(X, Y)

  # COnfigures the plot
  fig = plt.figure()
  ax = plt.axes()

  # Plots the special contour
  contours = ax.contourf(X, Y, Z, 20, cmap='RdGy')
  im = ax.imshow(Z, extent=[-3, 3, -3, 3], origin='lower',
            cmap='RdGy', alpha=0.5)
  
  # Plots the point (px, py) and configures its legend
  ax.plot([px], [py], 'yo', label='Point ('+str(px)+','+str(py)+')')
  ax.legend(loc='upper left', frameon=True, framealpha=0.9)

  # Shows a colorbar on the right side
  plt.colorbar(im, ax=ax)

  # Sets axis labels and title
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title(f.__name__)
  return
```

Plotting f1, f2, and f3 gives the following results:

```
plot_density_function(f1, px, py)
plot_density_function(f2, px, py)
plot_density_function(f3, px, py)
```
